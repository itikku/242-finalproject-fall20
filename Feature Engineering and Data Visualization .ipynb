## Focus on the ENERGY STAR Score and its trend

```python
data_not_null['ENERGY STAR Score'].value_counts()
```

```python
#Green: 75-100
#Yellow: 26-74
#Red: 0-25

######################### Creation of a new dataset with a column 'Label' ######################################
data_cat = data_not_null
label = []

for obs in data_cat['ENERGY STAR Score'] : 
    if obs < 25.0 : 
        label.append(0)
    elif obs >= 25.0 and obs < 74.0 :
        label.append(1)
    elif obs >= 74.0 : 
        label.append(2)


data_cat.insert(loc = 3, column = 'Label', value = label, allow_duplicates=False)
#print(len(category))
#print(len(data_categorized['ENERGY STAR Score']))
```

PIM Link is a useful feature because it links the parcel to the address and relevant information about the building. 
<br>

We may drop 'Label' as it's directly linked to ENERGY STAR Score, but we can also use it as the dependent variable for multi class classification. 


<br>
*As discussed, it's more relevant to use the 'source' types of data, the below explanation enforces .* 

* So we can keep 'Percent Better than National Median Source EUI', as it contributes to 4923 (1388 non NaN) rows VS 609 rows for 'Percent Better than National Median Site EUI'. 
* Similarly, we use 'Source EUI (kBtu/ft2)' (2448 obs). 
* We keep 'Weather Normalized Source EUI (kBtu/ft2)' (2448 obs). 
<br>
* Thus, we leave 'Percentage Better than National Median Source EUI' and 'Percent Better than National Median Site EUI', as they contribute to only 609 non NaN rows. 
* We leave 'Site EUI (kBtu/ft2)' (1503 obs). 
* We drop 'Weather Normalized Site EUI (kBtu/ft2)' (1508 obs). 

<br>

The feature 'Reason for Exemption' has no statistical significance on the prediction of the ENERGY Score. 

<br>

The features 'Property Type' and 'Property Type - Self Selected' are similar, but the latter is more varied. Question: if we do categorical variables, it might be easier to use Property Type.
